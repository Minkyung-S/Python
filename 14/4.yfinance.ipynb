```python
CPNG=yf.download('CPNG',start = '2020-01-01')
CPNG.head()
#쿠팡주가데이터
### 뒤에도 계속 데이터를 불러오기 위해 = 잊지말기 !!!!!
```

```python
##소수점 2번째 자리까지만 두고 반올림
CPNG[["Open","High","Low","Close"]]=\
    CPNG[["Open","High","Low","Close"]].round(2)
    
CPNG.head()
```

```python
## 결측치가 존재하는가??
CPNG.isna().sum()
```

```python
## 등락폭, 등락율 파생변수 생성
## 등락폭 = 종가-시가
## 등락율 = 등락폭 / 시가 * 100

CPNG["등락폭"]=CPNG["Close"]-CPNG["Open"]
CPNG["등락율"]=CPNG["등락폭"]/CPNG["Open"]*100
CPNG.head()
```

```python
## 시각화
import matplotlib.pyplot as plt
```

```python
plt.plot(CPNG['등락율'].values)
plt.show()
#스리즈는 인덱스값(아메리카노)과 value의 값(5,000)이 있음. -> value 값만으로 변경하고 싶다면??? 뒤에 .values하기!
```

```python
#그래프 4개를 출력
#subplot(행의수, 열의수, 선택할 영역)
# -> 영역을 나누는 작업, 나눈 영역에서 영역을 선택
# ex) subplot(2,2,1)-> 2행2열로 나눈다.->4개를 만든다->1번 영역 선택
plt.subplot(2,2,1)
plt.plot(CPNG['Open'].values)
plt.subplot(2,2,2)
plt.plot(CPNG['Close'].values)
plt.subplot(2,2,3)
plt.plot(CPNG['Volume'].values)
plt.subplot(2,2,4)
plt.plot(CPNG['등락율'].values)
plt.show()



```
