1.purchase_date 컬럼의 데이터를 시계열 데이터로 변경
2.item_name의 데이터를 모두 대문자로 변경
3.item_name의 데이터에서 공백을 삭제
4.item_name를 기준으로 오름차순 정렬
5.인덱스 초기화
6.결측치 없애기

```python
import pandas as pd
```

```python
uriage = pd.read_csv("../csv/uriage.csv")
```

```python
uriage.head()
```

```python
uriage.isna().sum()
```

```python
#1. purchase data 컬럼의 데이터를 시계열로 변경
# pd.to_datetime(변경할데이터,format = "")

uriage["purchase_date"]=pd.to_datetime(uriage["purchase_date"])
#이 데이터의 경우 datatime이 기본으로 잘 되어 있어서 뒤에 포멧값 필요없다. 
uriage.head()
```

```python
#2. item_name에 있는 데이터를 모두 대문자로 통일
##upper() 함수를 이용하면 모든 문자가 대문자로 변경
##스리즈를 문자열로 바꾼다 -> str
uriage["item_name"].str.upper()
```

```python
## 3. item_name의 데이터공백제거
#공백을 제거하는 함수
#strip() -> 양쪽의 공백을 제거하는 함수 () -> 글자와 글자 사이의 공백을 제거하지 못함
#replace()-> 문자열을 대체하는 함수-> .replace(" ","") 

#case1
uriage["item_name"].str.replace(" ","")

#case2
#apply 함수 -> 일일이 데이터 값을 입력하지 않고 불러올 수 있게함

def change(x):
    result = x.replace(" ","")
    return result

uriage["item_name"].apply(change)


#case3
#apply 함수 + lambda 함수
#함수를 사용하는 이유-> 반복된 작업을 할 때 코드의 라인이 줄어듬. 확인하기가 편하고 수정이 편해짐


## uriage["item_name"]= 을 해야지 뒤에 변수도 해당 식을 적용함
uriage["item_name"]= uriage["item_name"].apply(lambda x: x.replace(" ",""))


```

```python
##4.item_name을 기준으로 오름차순 정렬

#uriage.sort_values("item_name",inplace=True)=
uriage = uriage.sort_values("item_name")
uriage.head()
```

```python
#5. index 없애기
#uriage.reset_index().drop("index", axis=1) =

uriage.reset_index(drop=True, inplace=True)
```

### 6. 결측치의 데이터를 치환
1. 상품A의 item_price의 값은 결측치 중 상품이 A면 이 값을 치환
2. 나머지 상품들도 마찬가지로 같은 작업을 반복


```python
### 일부 데이터만 불러오기 
#uriage["item_name"]

## uriage.loc[행의 위치, 열의 위치]
## uriage.loc[[특정 값]]-> 특정 값이 있는 것만 출력

#uriage.loc[[0,1,2,3]]
```

```python
flg_null=uriage["item_price"].isna()

##스리즈의 값을 이 값에 담겠다.
#flg_null=
```

```python
#결측치인 값을 확인
uriage.loc[flg_null]
```

```python
##결측치가 아닌 값만 확인 
#~ 연산자 : not -> True- False 변경. 반대도 마찬가지로 행동
uriage.loc[~flg_null]
```

```python
#싱픔인 A인 데이터 프레임 출력
#uriage["item_name"] == "상품A" -> 이 경우는 시리즈 형태
uriage.loc[uriage["item_name"] == "상품A" ]
```

```python
##상품 A 이고 결측치인 데이터프레임만 추출
uriage.loc[(uriage["item_name"] == "상품A") & flg_null]
```

```python
## 상품A인 제품들의 결측치를 제외한 item_price의 평균값 구하기

# uriage.loc[(uriage["item_name"] == "상품A") & ~flg_null]["item_price"].mean() =
uriage.loc[(uriage["item_name"] == "상품A") & ~flg_null,"item_price"].mean()
# 행,열
```

```python
a_price = uriage.loc[(uriage["item_name"] == "상품A") & ~flg_null, "item_price"]
```

```python
## 상품A인 제품들의 결측치인 데이터프레임데이터만 출력
uriage.loc[(uriage["item_name"] == "상품A") & flg_null].fillna(a_price)
```
