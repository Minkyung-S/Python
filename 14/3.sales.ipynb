```python
import pandas as pd
```

### csv 파일 로드 데이터를 튜닝
1. csv 폴더에 있는 sales records.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit' 컬럼을 삭제
3. 컬럼의 이름 변경 ['권역', '국가', '상품종류', '채널', '우선순위', '주문일자', '발송일자', '단가', '원가']
4. 권역, 국가를 기준으로 오른차순 정렬
5. index의 값을 초기화하고 기존의 인덱스는 삭제

```python
sales = pd.read_csv("../csv/sales records.csv")
sales_df = sales #순서 바뀌면 안됨! (없는 값 = 있는 값)
sales_df.head()
```

```python
sales_df.drop(["Order ID","Total Revenue","Total Cost","Total Profit"], axis=1, inplace=True)
```

```python
sales_df.head()
```

```python
sales_df.columns = ["권역", "국가", "상품종류", "채널", "우선순위","주문일자", "발송일자", "판매단위", "단가", "원가"]
```

```python
sales_df.head()
```

```python
sales_df.sort_values(["권역","국가"], inplace=True)
```

```python
sales_df.head()
```

```python
sales_df.reset_index(drop=True, inplace=True)
```

```python
sales_df.head()
```

```python
##reset_index() 안에 drop은 기본값이 false
##drop 매개변수의 인자 값이 True가 되면 기존의 인덱스를 삭제하고 
#index를 초기화 한다
sales.sort_values(["국가"]).reset_index(drop=True)
#inplace를 하지 않아서 뒤에 값은 해당 값 고려 X
```

```python
##단가 컬럼의 float 형태의 데이터를 반올림-> int의 형태로 변경
#반올림 함수 .round(n)
#n의 값은 소수점의 자릿수
#sales_df["단가"].round(0)        #round(-1)은 맨뒷자리도 0
#데이터의 형태는 스리즈 values들의 데이터 형태는 float
#astype("type")-> astype("int")

sales_df["단가"]=sales_df["단가"].round(0).astype("int")
```

```python
sales_df.head()
```

```python
sales_df["원가"]=sales_df["원가"].astype("int")
#round()가 없기 때문에 반올림이 아닌 버림이 됨 -> 오차증가
```

```python
sales.head()
```

1. 새로운 파생변수 (총 단가, 총 원가, 총 이윤)생성
2. 총 단가 = 단가 * 판매단위
3. 총 원가 = 원가 * 판매단위
4. 총 이윤 = 총 단가 - 총 원가

```python
a = [1,2,3]
b = [4,5,6]
print(a+b)
```

```python
sales["총 단가"] = sales["단가"] * sales["판매단위"]
```

```python
sales["총 원가"] = sales["원가"] * sales["판매단위"]
```

```python

sales["총 이윤"] = sales["총 단가"] - sales["총 원가"]
```

```python
# 그룹화 함수 -> groupby(기준이 되는 컬럼)
test = sales.groupby("국가").sum()
```

```python
test[["총 이윤"]] #바깥 []을 하면 표처럼 나옴!
```

```python
## 총 이윤을 기반으로 내림차순

## sort_values()-> 기본 값은 오름차순 정렬
## 내림차순 정렬로 변경하려면 함수의 매개변수 ascending =
# True 오름차순 (기본값), False 내림차순

test.sort_values(["총 이윤"], ascending=False).head(10)
```
