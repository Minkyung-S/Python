```python
import matplotlib.pyplot as plt
```

```python
##plot 그래프 -> 라인 그래프 형성
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,200,3,500])
plt.show()
```

```python
x = [1,2,3,4]
y = [10,3,4,9]

#plot(x축값,y축값,옵션)
##옵션 선이 종류, 선의 색

plt.plot( x, y, "r")
plt.show()
```

```python
y2 = [4,2,5,1]
#첫번째 라인의 값
plt.plot(x,y,"--y")

#두번째 라인의 값
plt.plot(x,y2,"*g")

#x,y축 이름, 그래프 이름
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Class & Count") # 한글 안됨, 폰트추가 코드필요

#범뢰추가
plt.legend(["A","B"], loc=7)
plt.show()
```

```python
#subplot()-> 여러 개의 그래프를 그릴 때 영역을 나누고 영역을 지정한다.
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.show()
```

```python
#바형 그래프
x = [1,2,3,4]
y = [80,60,90,70]

plt.bar(x,y)
plt.show()
```

```python
##막대그래프 너비 변경, 막대의 색 변경
#bar형 그래프는 x,y 값 모두 들어가야 됨 (plot과 다름)
plt.bar(x, y, color=['y','g','b','r'],width=.4)
#x축 이름변경
plt.xticks(x,("maths", "science","history","english"))
plt.show()
```

```python
##산점도 그래프
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.show()
```

```python
#산점도 그래프 옵션
#양의 확장, 색상 변경, 투명도 조절

color = np.random.rand(50)
area= (20* np.random.rand(50))**2

plt.scatter(x,y,s = area,c= color,alpha=0.3)
```
